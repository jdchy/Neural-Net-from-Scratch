## Sum function

```python
def sum(inputs, weights):
    s = 0
    for i in range(len(inputs)):
        s += inputs[i] * weights[i]
    return s

def sum_2(inputs, weights):
    return inputs.dot(weights)
```

```python
s = sum_2(inputs,weights)
```

## Step function

```python
def step_function(sum):
    if (sum >= 1):
        return 1
    return 0
```
