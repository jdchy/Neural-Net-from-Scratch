### Load Dataset

```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
inputs = iris.data
inputs
```

```python
iris.feature_names
```

```python
outputs = iris.target
outputs
```

```python
iris.target_names
```

```python
inputs.shape
```

```python
outputs.shape
```

### Train and test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
X_train[0:5]
```

```python
X_test[0:5]
```

```python
y_train[0:5]
```

```python
y_test[0:5]
```

### Neural network

```python
network = MLPClassifier(max_iter=2000, 
                        verbose=True,
                        tol=0.0000100,
                        activation = 'logistic',
                        solver = 'adam',
                        learning_rate = 'constant',
                        learning_rate_init = 0.001,
                        batch_size = 32,
                        hidden_layer_sizes = (4, 5)
                        #early_stopping = True,
                        #n_iter_no_change = 50
                        )
network.fit(X_train, y_train)
```

```python
network.classes_
```

```python
network.coefs_
```

```python
network.intercepts_
```

```python
network.n_layers_
```

```python
network.n_outputs_
```

```python
network.out_activation_
```

### Neural network evaluation

```python
X_test.shape
```

```python
predictions = network.predict(X_test)
predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, predictions)
```

```python
cm = confusion_matrix(y_test, predictions)
cm
```
